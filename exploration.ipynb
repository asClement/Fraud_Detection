## Analyse bivariée

### operation vs fraud_flag

```python
sns.countplot(df, x='operation', hue='fraud_flag', order=['op_01', 'op_02', 'op_03', 'op_04', 'op_05'])
```

Remarque : Toutes les transactions frauduleuses se concentrent dans la catégorie op_3. Première piste pour feature engineering

### operation vs amount

```python
fig, ax=plt.subplots(1,2,figsize=(15,6))
grps = df.groupby('operation')
sns.barplot(data=grps['amount'].agg(pd.Series.mean), ax=ax[0])
sns.boxplot(df, x='operation', y='amount', order=['op_01', 'op_02', 'op_03', 'op_04', 'op_05'], ax=ax[1])
```

Note : Les transactions frauduleuses ont des montants moyens (≃70 K de la classe op_3) le plus souvent

### period vs amount

```python
fig, ax= plt.subplots(1, 2, figsize=(15,6))
axs= ax.flatten()
grps= df.groupby('period')
data=grps['amount'].agg(pd.Series.std)
sns.barplot(data=grps['amount'].agg(pd.Series.sum), ax=axs[0])
sns.barplot(data=grps['amount'].agg(pd.Series.mean), ax=axs[1])
```

Remarque : La structure en 3 pics observées au niveau de la courbe des volumes de transactions par `period` est également respectée avec les volumes monétaires par période. Une intuition sort également de ce graphique : en se basant sur le graphe précédent de la moyenne des coûts par type d'opération, on peut **'imaginer'** une concentration des types d'opération selon les périodes.

### period vs operation

```python
fig, ax= plt.subplots(2, 3, figsize=(25,8))
axs= ax.flatten()

data= df[df['operation']=='op_01']
sns.countplot(data, x='period', ax=axs[0])
axs[0].set_title('op_01')

data= df[df['operation']=='op_02']
sns.countplot(data, x='period', ax=axs[1])
axs[1].set_title('op_02')

data= df[df['operation']=='op_03']
sns.countplot(data, x='period', ax=axs[2])
axs[2].set_title('op_03')

data= df[df['operation']=='op_04']
sns.countplot(data, x='period', ax=axs[3])
axs[3].set_title('op_04')

data= df[df['operation']=='op_05']
sns.countplot(data, x='period', ax=axs[4])
axs[4].set_title('op_05')

sns.countplot(df[df['fraud_flag']==1], x='period', ax=axs[5])
axs[5].set_title('Count des period pour fraud_flag=1')
plt.tight_layout()
plt.show()
```

L'intuition de la partie précédente constitue à présent une piste pour le feature engineering. On remarque notamment une périodicité très similaire entre les opérations de type `op_03` et `op_05`. Des périodes n’ont des transactions que de type `op_03` et `op_05`. On peut supposer ceci : **Une transaction potentiellement frauduleuse est précédée d’une petite transaction**. \
Les 3 pics d'activité repérés plus haut se justifient par la présence des autres types d'activités à ces intervalles (7 à 26, 34 à 49, 93 à 97).

## Explorations complémentaires

On remarque des comptes d'origine et de destination avec des soldes négatifs dans le `describe()`. Vérifions si c'est un signe de fraude

```python
print("=====Répartition de la fraude pour les soldes d'origine négatifs avant transaction=====")
print()
print(df[df['origin_balance_before']<0.0]['fraud_flag'].value_counts())
print()
print("=====Répartition de la fraude pour les soldes d'origine négatifs après transaction=====")
print()
print(df[df['origin_balance_after']<0.0]['fraud_flag'].value_counts())
print()
print("=====Répartition de la fraude pour les soldes d'origine négatifs après transaction et avec solde de départ négatifs=====")
print()
print(df[(df['origin_balance_before']<0.0) & (df['origin_balance_after']<0.0)]['fraud_flag'].value_counts())
print()
print("=====Répartition de la fraude pour les soldes de destination négatifs avant transaction=====")
print()
print(df[df['destination_balance_before']<0.0]['fraud_flag'].value_counts())
print()
print("=====Répartition de la fraude pour les soldes de destination négatifs après transaction=====")
print()
print(df[df['destination_balance_after']<0.0]['fraud_flag'].value_counts())
print()
print("=====Répartition de la fraude pour les soldes de destination négatifs après transaction et avec solde de départ négatifs=====")
print()
print(df[(df['destination_balance_before']<0.0) & (df['destination_balance_after']<0.0)]['fraud_flag'].value_counts())
```

Une transaction frauduleuse est toujours destinée à un compte ayant un solde positif (>=0.0).

## Exploration de pistes pour feature engineering

```python
df['log_amount']= np.log1p(df['amount'])
```

```python
fig, ax= plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(df, x='fraud_flag', y='log_amount', ax=ax[0])
sns.barplot(df.groupby('fraud_flag')['log_amount'].agg(pd.Series.mean), ax=ax[1])
```

```python
print("=====Statistiques descriptives de log_amount pour fraud_flag = 0=====")
print()
print(df[df['fraud_flag']==0]['log_amount'].describe())
print()
print("=====Statistiques descriptives de log_amount pour fraud_flag = 1=====")
print()
print(df[df['fraud_flag']==1]['log_amount'].describe())
```

`log_amount` vient confirmer la concentration des transactions frauduleuses dans une plage bien précise, contrairement aux transactions légitimes.

```python
df['is_dest_positive']= df['destination_balance_before'].apply(lambda x: 1 if x>=0.0 else 0)
pd.crosstab(df['is_dest_positive'], df['fraud_flag'])
```

```python
df['is_op_3']= (df['operation']=='op_03').astype(int)
sns.countplot(df, x='is_op_3', hue='fraud_flag')
```

```python
df['amount_orig_balance_before_ratio']= np.abs((df['origin_balance_before'] / df['amount']).round(4))
fig, ax= plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(df, x='fraud_flag', y='amount_orig_balance_before_ratio', ax=ax[0])
sns.barplot(df.groupby('fraud_flag')['amount_orig_balance_before_ratio'].agg(pd.Series.mean), ax=ax[1])
```

```python
print('=====Statistiques de amount_orig_balance_before_ratio pour fraud_flag = 0')
print(df[df['fraud_flag']==0]['amount_orig_balance_before_ratio'].describe())
print()
print('=====Statistiques de amount_orig_balance_before_ratio pour fraud_flag = 1')
print(df[df['fraud_flag']==1]['amount_orig_balance_before_ratio'].describe())
```

```python
df['amount_dest_balance_after_ratio']= np.abs((df['destination_balance_after'] / df['amount']).round(4))
fig, ax= plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(df, x='fraud_flag', y='amount_dest_balance_after_ratio', ax=ax[0])
sns.barplot(df.groupby('fraud_flag')['amount_dest_balance_after_ratio'].agg(pd.Series.mean), ax=ax[1])
```

```python
print('=====Statistiques de amount_dest_balance_after_ratio pour fraud_flag = 0')
print(df[df['fraud_flag']==0]['amount_dest_balance_after_ratio'].describe())
print()
print('=====Statistiques de amount_dest_balance_after_ratio pour fraud_flag = 1')
print(df[df['fraud_flag']==1]['amount_dest_balance_after_ratio'].describe())
```
